# Healthcare Encounter Analytics

Exploration, modeling, and interpretation for the cleaned healthcare dataset.

```python
import pandas as pd
from pathlib import Path

df = pd.read_csv('healthcare_dataset_cleaned.csv')
df.head()
```

```python
df.shape, df.isna().sum().sort_values(ascending=False).head(10)
```

```python
df['medical_condition'].value_counts()
```

```python
df.groupby('medical_condition').agg(encounters=('encounter_id','count'), avg_billing=('billing_amount','mean'), avg_los_days=('length_of_stay_days','mean')).sort_values('avg_billing', ascending=False)
```
